```python
# TODO: Ask user to enter a score (0-100) and print the grade

score = int(input("Enter your score"))

if score < 0 or score > 100:
    print("Invalid score")
elif score >= 90:
    print("Your Grade is A")
elif score >= 80:
    print("Your Grade is B")
elif score >= 70:
    print("Your Grade is C")
elif score >= 60:
    print("Your Grade is D")
else:
    print("Your Grade is F")
```

## 🔸 Step 5: Tuples, Sets, and Dictionaries
Explore basic operations on key Python data structures.

```python
# Create a tuple of 5 items
# tuples are immutable.Once you created you cannot change, add or remove anything from the tuple.
sample_tuple = ("tomato", "potato", "carrot", "garlic", "ginger")
print("tuple", sample_tuple)

# count() method : Count how many times an item appears
item_count = sample_tuple.count("carrot")
print("item count :",item_count)

# index() method : find the position of item
garlic_index = sample_tuple.index("garlic")
print("garlic index :", garlic_index)

# Create a set and add values
# A set keeps only unique items and ignores duplicates. It does not track the order of elements.
sample_set = set()
sample_set.add("apple")
sample_set.add("banana")
sample_set.add("orange")
sample_set.add("pineapple")

sample_set2 = {"mango","grapes","banana", "stawberry" }

# union method : combine items from both set and remove duplicates
union_method = sample_set.union(sample_set2)
print("union method :" , union_method)

# intersection method : returns common elements
intersection_method = sample_set.intersection(sample_set2)
print("intersection method :", intersection_method)



# Create a dictionary
sample_dict = {'name': 'Alice', 'age': 25}

# Print the dictionary
print("Original Dictionary:", sample_dict)

# Access a value
print("Name:", sample_dict["name"])
print("Age:", sample_dict["age"])

# Add a new key-value pair
sample_dict["city"] = "Lahore"
print("After adding city:", sample_dict)

# Update a value
sample_dict["age"] = 26
print("After updating age:", sample_dict)

# Remove a key-value pair
sample_dict.pop("city")
print("After removing city:", sample_dict)

# Get all keys
print("Keys:", sample_dict.keys())

# Get all values
print("Values:", sample_dict.values())

# Get all key-value pairs
print("Items:", sample_dict.items())

# Count the number of key-value pairs
print("Length:", len(sample_dict))

```

## 🔸 Step 6: Challenge - Word Frequency Counter
Write a Python program that counts the frequency of each word in a string.

```python
# TODO: Word frequency counter using dictionary


text = "Python is simple yet powerful. Python is easy to learn."


words = text.lower().replace(".", "").split()

word_freq = {}

for word in words:
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1

print(word_freq)
```
